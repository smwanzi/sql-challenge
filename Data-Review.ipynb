# Data Modelling

In this assignment, we are asked to research employees data of Pewlett Hackard corporation from the 1980s and 1990s. Using the CSV files provided, design the tables to hold data in the CSVs, import the CSVs into a SQL database, and answer questions about the data.

The CSV datasets provided:

* departments
* dept_emp
* dept_manager
* employees
* salaries
* titles

## ERD
Inspect the CSVs and sketch out an ERD of the tables.

## Relationships

A review of the CSV files shows that you could designate two main tables: departments and employees and relate the other tables to these two.

departments - has dept_no as a Primary Key (PK) 

employees - has emp_no primary key  (PK)

The other four tables have at least one of those two fields as foreign keys.



## Data Types

All *date columns defined as DATE type

Variable character columns set to VARCHAR

emp_no set to INTEGER

dept_no set to CHAR(4) since dept_no values are complex 

gender set to CHAR(1) since value is either 'M' or 'F'
